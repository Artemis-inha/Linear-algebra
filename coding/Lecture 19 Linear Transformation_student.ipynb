Transformation in Plane

Example 1:  Stretch

$A = \begin{bmatrix} 2 & 0 \\ 0 & 1 \end{bmatrix}$

$A = \begin{bmatrix} 1 & 0 \\ 0 & 2 \end{bmatrix}$

$A = \begin{bmatrix} 2 & 0 \\ 0 & 2 \end{bmatrix}$


```python
%matplotlib inline
from math import pi, sin, cos
import matplotlib.pyplot as plt
import numpy as np

coords = np.array([[0,0],[1,0.5],[1,1.5],[0,1],[0,0]])
coords = coords.transpose()
coords
x = coords[0,:]
y = coords[1,:]

A = np.array([[1,0],[0,2]])
A_coords = A@coords

x_LT1 = A_coords[0,:]
y_LT1 = A_coords[1,:]

# Create the figure and axes objects
fig, ax = plt.subplots()

# Plot the points.  x and y are original vectors, x_LT1 and y_LT1 are images
ax.plot(x,y,'ro')
ax.plot(x_LT1,y_LT1,'bo')

# Connect the points by lines
ax.plot(x,y,'r',ls="--")
ax.plot(x_LT1,y_LT1,'b')

# Edit some settings
ax.axvline(x=0,color="k",ls=":")
ax.axhline(y=0,color="k",ls=":")
ax.grid(True)
ax.axis([-1,4,-1,4])
ax.set_aspect('equal')
ax.set_title("Horizontal Stretch");
```

Example 2: Reflection

Example 3: Rotation

$A = \begin{bmatrix} \cos(\theta) & -\sin(\theta) \\ \sin(\theta) & \cos(\theta) \end{bmatrix}$

```python
theta = pi/2
R = np.array([[cos(theta),-sin(theta)],[sin(theta),cos(theta)]])
R_coords = R@coords

x_LT3 = R_coords[0,:]
y_LT3 = R_coords[1,:]

# Create the figure and axes objects
fig, ax = plt.subplots()

# Plot the points.  x and y are original vectors, x_LT1 and y_LT1 are images
ax.plot(x,y,'ro')
ax.plot(x_LT3,y_LT3,'bo')

# Connect the points by lines
ax.plot(x,y,'r',ls="--")
ax.plot(x_LT3,y_LT3,'b')

# Edit some settings
ax.axvline(x=0,color="k",ls=":")
ax.axhline(y=0,color="k",ls=":")
ax.grid(True)
ax.axis([-2,2,-1,2])
ax.set_aspect('equal')
ax.set_title("Rotation");
```

Example 5 Translation

x= x+ 3,  y = y + 1


$A = \begin{bmatrix} 1 & 0 & 3 \\ 0 & 1 & 1 \\ 0 & 0 & 1 \end{bmatrix}$


```python
coords = np.array([[0,0],[0,3],[1,3],[1,1],[2,1],[2,0],[0,0]])
coords = coords.transpose()
x = coords[0,:]
y = coords[1,:]
Ones = np.ones((1,7))
coords = np.vstack((x,y,Ones))

S = np.array([[1,0, 3],[0,1, 1], [0, 0 , 1]])

S_coords = S@coords
x_LT5 = S_coords[0,:]
y_LT5 = S_coords[1,:]

# Create the figure and axes objects
fig, ax = plt.subplots()

# Plot the points.
ax.plot(x,y,'ro')
ax.plot(x_LT5,y_LT5,'bo')

# Connect the points by lines
ax.plot(x,y,'r',ls="--")
ax.plot(x_LT5,y_LT5,'b')

# Edit some settings
ax.axvline(x=0,color="k",ls=":")
ax.axhline(y=0,color="k",ls=":")
ax.grid(True)
ax.axis([-1,6,-1,6])
ax.set_aspect('equal')
ax.set_title("Translation");
```

3D Linear transformation

```python
import plotly.graph_objects as go


# 구형 point cloud 생성
np.random.seed(0)
phi = np.random.uniform(0.3*np.pi, 0.8*np.pi, 500)
theta = np.random.uniform(0.2*np.pi, 0.7*np.pi, 500)
r = 1.0

x = r * np.sin(phi) * np.cos(theta)
y = r * np.sin(phi) * np.sin(theta)
z = r * np.cos(phi)

points = np.vstack([x, y, z]).T  # shape: (500, 3)


def rotation_matrix_xyz(rx, ry, rz):
    Rx = np.array([
        [1, 0, 0],
        [0, np.cos(rx), -np.sin(rx)],
        [0, np.sin(rx), np.cos(rx)]
    ])
    Ry = np.array([
        [np.cos(ry), 0, np.sin(ry)],
        [0, 1, 0],
        [-np.sin(ry), 0, np.cos(ry)]
    ])
    Rz = np.array([
        [np.cos(rz), -np.sin(rz), 0],
        [np.sin(rz), np.cos(rz), 0],
        [0, 0, 1]
    ])
    return Rz @ Ry @ Rx  # 순서: X → Y → Z 회전

# 회전 각도 (라디안 단위)
rx, ry, rz = np.radians([90, 90, -90])
R = rotation_matrix_xyz(rx, ry, rz)
rotated_points = (R @ points.T).T



fig = go.Figure()

# 원본 (빨간색)
fig.add_trace(go.Scatter3d(
    x=points[:,0], y=points[:,1], z=points[:,2],
    mode='markers',
    marker=dict(size=2, color='red'),
    name='Original'
))

# 회전된 (파란색)
fig.add_trace(go.Scatter3d(
    x=rotated_points[:,0], y=rotated_points[:,1], z=rotated_points[:,2],
    mode='markers',
    marker=dict(size=2, color='blue'),
    name='Rotated'
))

fig.update_layout(
    title=f"3D Point Cloud Rotation around Z-axis)",
    scene=dict(
        xaxis=dict(range=[-1, 1]),
        yaxis=dict(range=[-1, 1]),
        zaxis=dict(range=[-1, 1]),
        aspectmode='cube'),
    width=800,
    height=600,
)

fig.show()
```

Application: Image transformation

```python
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt
import torchvision.transforms.functional as TF
from PIL import Image
import numpy as np
from torchvision.transforms import InterpolationMode
```

```python
mtx = np.array([[1, 0.7, 0.4],
                [0.9, 0.6, 0.3],
                [0.8, 0.5, 0.3]])

# Plot transformation matrix
plt.imshow(mtx, cmap='gray')
plt.axis('off')
plt.show()
```

```python
data = fetch_lfw_people(min_faces_per_person=30, resize=1)
images = data.images

index = 10
plt.imshow(images[index], cmap='gray')
plt.title(data.target_names[data.target[index]])
plt.axis('off')
plt.show()
```

```python
print(len(images))
print(images[0])
print(images.shape)
```

```python
img_array = images[0] *255
img = Image.fromarray(img_array.astype(np.uint8), mode='L')

transformed = TF.affine(img, angle=30, translate=[0, 0], scale=1, shear=[0, 0], interpolation=InterpolationMode.BILINEAR)
# transformed = TF.affine(img, angle=30, translate=[0, 0], scale=1, shear=[0, 0], interpolation=InterpolationMode.NEAREST)

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(transformed, cmap='gray')
plt.title("Affine Transformed")
plt.axis('off')
plt.show()
```
