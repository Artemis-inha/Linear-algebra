Least square using the library

```python
# data generation
np.random.seed(0)
x = np.linspace(-3, 3, 100)
y_true = 2*x**2 - 4*x + 5
noise = np.random.normal(scale=0.1, size=x.size)
y = y_true + noise # add normal distribution noise

# fitting using the library
coeffs = np.polyfit(x, y, deg=2)
# print(coeffs)
poly = np.poly1d(coeffs)
print(poly)

# plot
plt.scatter(x, y, alpha=0.4)
plt.plot(x, poly(x), color='red')
plt.grid(True)
plt.show()
```

```python
def polyfit(x, y, deg):
  data_num = x.size
  A = np.zeros([data_num, deg+1])
  b = y.T
  A[:, 0] = 1
  for i in range(deg):
    A[:, i+1] = x**(i+1)
    A[:, i+1] = np.power(x, i+1)

  coeffs = np.linalg.inv(A.T@A)@A.T@b
  coeffs = np.flip(coeffs)
  return coeffs

print(coeffs)
```

```python
def fit_circle(x, y):
    x = np.asarray(x).reshape(-1)
    y = np.asarray(y).reshape(-1)
    A = np.column_stack((x, y, np.ones(x.size)))
    b = -(x**2 + y**2)

    # Least squares solution to A @ [D, E, F] = b

    w = 0.0001
    coeffs= np.linalg.inv(A.T@A - w*np.eye(3))@A.T@b
    D, E, F = coeffs[0], coeffs[1], coeffs[2]

    # Recover circle parameters
    a = -D / 2
    b = -E / 2
    r = np.sqrt(a**2 + b**2 - F)

    return a, b, r
```
